
# Assignment.2

![](2.png)

Реализовать многопоточный, однопоточный, многопроцесный варианты. Сравнить результаты.

# Assignment.3

* В папке FILES лежит архив trades_info.zip. Распаковать его с помощью python
* В этой папке лежит информация о сделках на бирже по ценным бумагам.  
* Ценные бумаги выпускаются партиями, в которых может содержаться большое количество единиц ценных бумаг.
* Каждая  партия имеет свой торговый код на бирже - тикер.  
* При этом, все ценные бумаги из этой партии одинаковы в цене, поэтому говорят о цене одной бумаги.
* У разных выпусков бумаг - разные цены, которые могут отличаться в сотни и тысячи раз.
* Каждая биржевая сделка характеризуется:
*   тикер ценнной бумаги
*   время сделки
*   цена сделки
*   обьем сделки (сколько ценных бумаг было куплено)
*
* В ходе торгов цены сделок могут со временем расти и понижаться. Величина изменения цен называтея волатильностью.
* Например, если бумага №1 торговалась с ценами 11, 11, 12, 11, 12, 11, 11, 11 - то она мало волатильна.
* А если у бумаги №2 цены сделок были: 20, 15, 23, 56, 100, 50, 3, 10 - то такая бумага имеет большую волатильность.
* Волатильность можно считать в виде
* отклонения в процентах от средней цены за торговую сессию:
*   средняя цена = (максимальная цена + минимальная цена) / 2
*   волатильность = ((максимальная цена - минимальная цена) / средняя цена) * 100%
* Например для бумаги №1:
*   average_price = (12 + 11) / 2 = 11.5
*   volatility = ((12 - 11) / average_price) * 100 = 8.7%
* Для бумаги №2:
*   average_price = (100 + 3) / 2 = 51.5
*   volatility = ((100 - 3) / average_price) * 100 = 188.34%

* Необходимо вычислить 3 тикера с максимальной и 3 тикера с минимальной волатильностью.
* Бумаги с нулевой волатильностью вывести отдельно.
* Результаты вывести на консоль в виде:
*   Максимальная волатильность:
*       ТИКЕР1 - ХХХ.ХХ %
*       ТИКЕР2 - ХХХ.ХХ %
*       ТИКЕР3 - ХХХ.ХХ %
*   Минимальная волатильность:
*       ТИКЕР4 - ХХХ.ХХ %
*       ТИКЕР5 - ХХХ.ХХ %
*       ТИКЕР6 - ХХХ.ХХ %
*   Нулевая волатильность:
*       ТИКЕР7, ТИКЕР8, ТИКЕР9, ТИКЕР10, ТИКЕР11, ТИКЕР12
* Волатильности указывать в порядке убывания. Тикеры с нулевой волатильностью упорядочить по имени.
* В каждом файле в папке trades_info содержится данные по сделакам по одному тикеру, разделенные запятыми.
*   Первая строка - название колонок:
*       SECID - тикер
*       TRADETIME - время сделки
*       PRICE - цена сделки
*       QUANTITY - количество бумаг в этой сделке
*   Все последующие строки в файле - данные о сделках
<br>
**Реализовать задачу однопоточным, многопоточным и многопроцессным способами. Сравнить время выполнения и результаты**
